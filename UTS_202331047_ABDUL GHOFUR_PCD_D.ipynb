## SOAL PERTAMA

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("SOAL1.jpg")
```

```python
# Memisahkan channel BGR
B, G, R = cv2.split(image)

# Konversi BGR ke RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Membuat subplot untuk menampilkan gambar
fig, axs = plt.subplots(2, 2, figsize=(15, 8))

# Citra RGB
axs[0, 0].imshow(image_rgb)
axs[0, 0].set_title("CITRA KONTRAS")
axs[0, 0].axis('on')       # Menampilkan axis ticks
axs[0, 0].grid(False)      # Nonaktifkan grid

# Channel Biru
axs[0, 1].imshow(B, cmap='gray')
axs[0, 1].set_title("BIRU")
axs[0, 1].axis('on')
axs[0, 1].grid(False)

# Channel Merah
axs[1, 0].imshow(R, cmap='gray')
axs[1, 0].set_title("MERAH")
axs[1, 0].axis('on')
axs[1, 0].grid(False)

# Channel Hijau
axs[1, 1].imshow(G, cmap='gray')
axs[1, 1].set_title("HIJAU")
axs[1, 1].axis('on')
axs[1, 1].grid(False)

plt.tight_layout()
plt.show()
```

```python
# Menampilkan histogram dari RGB dan masing-masing channel
plt.figure(figsize=(20, 5))

plt.subplot(141)
plt.title('Histogram Asli')
plt.hist(image_rgb.ravel(), bins=256, range=[0, 256], color='gray')

plt.subplot(142)
plt.title('Histogram merah')
plt.hist(R.ravel(), bins=256, range=[0, 256], color='r')

plt.subplot(143)
plt.title('Histogram hijau')
plt.hist(G.ravel(), bins=256, range=[0, 256], color='g')

plt.subplot(144)
plt.title('Histogram biru')
plt.hist(B.ravel(), bins=256, range=[0, 256], color='b')

plt.tight_layout()
plt.show()
```

```python
# Load an image
color_image = cv2.imread('SOAL1.jpg')

# Convert from BGR to RGB for correct display in matplotlib
color_image_rgb = cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB)

# Convert to HSV for color detection
hsv = cv2.cvtColor(color_image, cv2.COLOR_BGR2HSV)

# Define blue range in HSV
lower_blue = np.array([100, 50, 50])
upper_blue = np.array([130, 255, 255])

# Define red ranges in HSV (red wraps around in HSV, so we need two ranges)
lower_red1 = np.array([0, 50, 50])
upper_red1 = np.array([10, 255, 255])
lower_red2 = np.array([170, 50, 50])
upper_red2 = np.array([180, 255, 255])

# Create masks
mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)
mask_red1 = cv2.inRange(hsv, lower_red1, upper_red1)
mask_red2 = cv2.inRange(hsv, lower_red2, upper_red2)

# Combine the two red masks
mask_red = cv2.bitwise_or(mask_red1, mask_red2)

# Create combined red-blue mask
mask_combined = cv2.bitwise_or(mask_blue, mask_red)

# Create plots with just two panels
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Show original image (in RGB)
axs[0].imshow(color_image_rgb)
axs[0].set_title('Citra Asli')
axs[0].axis('off')  # Hide axes

# Show combined red-blue mask
axs[1].imshow(mask_combined, cmap='gray')
axs[1].set_title('Red-Blue')
axs[1].axis('off')  # Hide axes

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

## SOAL KETIGA

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Membaca gambar asli dari file
path = "SOAL2.jpg"
img = cv2.imread(path)

# Konversi BGR ke RGB untuk ditampilkan dengan matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Konversi ke grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Pencerahan gambar gray 
bright_gray = cv2.convertScaleAbs(img_gray, alpha=1, beta=50)  # beta menambah kecerahan

# Penambahan kontras pada gambar gray
contrast_gray = cv2.convertScaleAbs(img_gray, alpha=1.5, beta=0)  # alpha meningkatkan kontras

# Untuk Kombinasi Dipercerah + Diperkontras
bright_contrast_gray = cv2.convertScaleAbs(img_gray, alpha=1.5, beta=50)

plt.figure(figsize=(20, 30))  # Lebar diperbesar

plt.subplot(6, 1, 1)
plt.imshow(img_rgb)
plt.title("Gambar Asli", fontsize=18)
plt.axis("off")

plt.subplot(6, 1, 2)
plt.imshow(img_gray, cmap='gray')
plt.title("Gambar Gray", fontsize=18)
plt.axis("off")

plt.subplot(6, 1, 3)
plt.imshow(bright_gray, cmap='gray')
plt.title("Gambar Gray yang Dipercerah", fontsize=18)
plt.axis("off")

plt.subplot(6, 1, 4)
plt.imshow(contrast_gray, cmap='gray')
plt.title("Gambar Gray yang Diperkontras", fontsize=18)
plt.axis("off")

plt.subplot(6, 1, 5)
plt.imshow(bright_contrast_gray, cmap='gray')
plt.title("Gambar Gray Dipercerah & Diperkontras", fontsize=18)
plt.axis("off")

plt.subplots_adjust(hspace=0.4)  # Atur jarak antar gambar
plt.show()
```
